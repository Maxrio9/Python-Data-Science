```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
gdp_life = pd.read_csv("all_data.csv")

print(gdp_life.Country.unique())
gdp_life.describe(include = "all")
```

```python
chile = gdp_life[gdp_life.Country == "Chile"]
china = gdp_life[gdp_life.Country == "China"]
germany = gdp_life[gdp_life.Country == "Germany"]
mexico = gdp_life[gdp_life.Country == "Mexico"]
usa = gdp_life[gdp_life.Country == "United States of America"]
zimbabwe = gdp_life[gdp_life.Country == "Zimbabwe"]
```

```python
for country in [chile, china, germany, mexico, usa, zimbabwe]:
    sns.scatterplot(data = country, x = "Life expectancy at birth (years)", y = "GDP")
    plt.title(country.Country.unique())
    
    plt.show()
    plt.clf()
```

```python
for country in [chile, china, germany, mexico, usa, zimbabwe]:
    print(country.Country.unique())
    print(country[["Life expectancy at birth (years)", "GDP"]].corr())
```

### Conclusion

Within every country we can see a very strong positive correlation between life expectancy at birth and GDP, the strongest being the US and the weakest China
